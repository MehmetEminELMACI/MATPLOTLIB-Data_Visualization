```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#This way, jupyter appears on the notebook.
```

```python
x=np.arange(20) # 0-19
y=np.random.normal(10,1,20) # average=10 standard deviation=1 20 adet sayÄ±
z=np.random.normal(10,2,20)
```

```python
y
```

```python
z
```

# PLOT

```python
plt.plot(x,y,color="black")
```

```python
fig=plt.figure() #Create figÃ¼r
ax=fig.add_subplot(111)
ax.plot(x,y,color="black",label="Thing #1")
ax.plot(x,z,color="red",label="Thing #2")
ax.legend()  #Change of X with respect to Y
```

# Add Labels, Styles / Change Color, Marker, Fontsize / Save the Plot / 
# Quality improvement

```python
#LABELS AND STYLES
fig=plt.figure() #FigÃ¼r oluÅtur
ax=fig.add_subplot(111)
ax.plot(x,y,marker="*",c="#FF0000")
ax.set_ylabel("$$$ Money",fontsize=10)
ax.set_xlabel("Years",fontsize=20)
fig.savefig("figure.png",dpi=300) 
#Save and quality improvement with dpi
```

# MULTIPLE PLOTS

```python
fig2=plt.figure(figsize=(15,5))
a=fig2.add_subplot(131)
a.scatter(x,y)
a.set_title("Scatter plot")

b=fig2.add_subplot(132)
b.hist(y)
b.set_title("Histogram")

c=fig2.add_subplot(133)
c.boxplot([y,x])
c.set_title("Box plot")

fig2.suptitle("Visualazition")
```
